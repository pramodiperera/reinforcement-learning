# Frozen Lake

***Problem statement*** - Frozen lake is an environment composed of tiles. The agent has to move from the intial tile to the Goal state. Tiles can be a safe frozen lake(F) or a hole(H) that gets you stuck. There are 4 possible actions left, right, up and down. The agent must learn to avoid holes in order to reach the goal in a minimum number of actions. further, each tile is represented by a letter in the environment's code.

```python
import gym
import random
import numpy as np
import time
from IPython.display import clear_output
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
env = gym.make("FrozenLake-v1", is_slippery=False)
env.reset()
env.render()
```

```python
# initializing Q-table
num_states = env.observation_space.n
num_actions = env.action_space.n

qtable = np.zeros((num_states, num_actions))
qtable
```

```python
'''
#### Values for each actions 

left - 0
down - 1
right - 2
up - 3

'''

# random action
action = env.action_space.sample()
print("action = ", action)

new_state, reward, done, info = env.step(action)
print("reward", reward)              # getting the reward(1) only when the agent reaches to the goal. In all other cases, 0 reward
```

### ***Q Learning***

```python
# 1. choose a random action, if the values for the state are zero. Otherwise take the max valued action
# 2. take the action
# 3. Update the Q
# do this until the agent reaches the Goal or get stucked in a hole. Then restart the env to strat a new episode


# initialize the q table
qtable = np.zeros((env.observation_space.n, env.action_space.n))

# hyperparameters
episodes = 1000
alpha = 0.5         # learning rate
gamma = 0.9         # discount rate
                    # In FrozenLake, we need a high gamma value, as we get the reward only at the end(when reaching to the goal)

goal_reached = []

print("Q table before training :", qtable)

for _ in range(episodes):
    state = env.reset()
    done = False
    goal_reached.append("failure")   #as a default, we set the outcome as a failure
    
    #until the agent reaches to the goal, or gets stuck in a hole
    while not done:
        if np.max(qtable[state]) > 0:
            action = np.argmax(qtable[state])    # argmax gives the index
        else:
            action = env.action_space.sample()
        
        # action
        new_state, reward, done, info = env.step(action)
        # update q
        qtable[state, action] = qtable[state, action] + alpha*(reward + gamma*np.max(qtable[new_state]) - qtable[state, action])
        # update the current state
        state = new_state
        
        if reward:
            goal_reached[-1] = "success"

            
print("\nQ table after training :", qtable)
    
# plot if the agent reached the goal in each episode
plt.figure(figsize=(12, 5))
plt.xlabel("episodes")
plt.ylabel("If the agent reached the goal")
ax = plt.gca()
ax.set_facecolor('#efeeea')
plt.bar(range(len(goal_reached)), goal_reached, color="#0A047A", width=1.0)
plt.show()
```

Now we have the updated Q table. Lets see how the agent perform.

```python
##### testing

test_episodes = 100
num_success = 0

for _ in range(test_episodes):
    state = env.reset()
    done = False
    
    while not done:
        if np.max(qtable[state]) > 0:
            action = np.argmax(qtable[state]) 
        else:
            action = env.action_space.sample()
        
        new_state, reward, done, info = env.step(action)
        state = new_state
        
        num_success += reward

print("Success rate = {} %".format(num_success/test_episodes*100) )
```

in each episode, the agent has reached to the Goal.
 
Below code shows the sequence of action the agent took to reach the goal. We'll see it for one episode.

```python
state = env.reset()
done = False
seq = []

while not done:
        if np.max(qtable[state]) > 0:
            action = np.argmax(qtable[state]) 
        else:
            action = env.action_space.sample()
        
        seq.append(action)
        new_state, reward, done, info = env.step(action)
        state = new_state
        
        clear_output(wait=True)
        env.render()
        time.sleep(0.5)

print("squence of actions= ", seq)
```

### ***Epsilon-Greedy Algorithm***

In above used method, the agent always use the highest q value (Exploitation). But there may be other better action than the one agent select. The agent has to take the ramdom action as well (Exploration). Therefore, we have to balance the exploitation and the exploration.

```python
######## Training

qtable = np.zeros((env.observation_space.n, env.action_space.n))

episodes = 1000
alpha = 0.5       
gamma = 0.9  
epsilon = 1               # amount of randomness
epsilon_decay = 0.001     # linear decay - fixed amount is decreased

goal_reached = []

print("Q table before training :", qtable)

for _ in range(episodes):
    state = env.reset()
    done = False
    goal_reached.append("failure")   
    
    #until the agent reaches to the goal, or gets stuck in a hole
    while not done:
        random_num = np.random.random()
        
        if random_num > epsilon:
            action = np.argmax(qtable[state])    # exploitation
        else:
            action = env.action_space.sample()   # exploration
        
        # action
        new_state, reward, done, info = env.step(action)
        # update q
        qtable[state, action] = qtable[state, action] + alpha*(reward + gamma*np.max(qtable[new_state]) - qtable[state, action])
        # update the current state
        state = new_state
        
        if reward:
            goal_reached[-1] = "success"
    
    # linear decay - reduce the epsilon
    epsilon = max(epsilon - epsilon_decay, 0)
    
            
        
print("\nQ table after training :", qtable)
    
# plot if the agent reached the goal in each episode
plt.figure(figsize=(12, 5))
plt.xlabel("episodes")
plt.ylabel("If the agent reached the goal")
ax = plt.gca()
ax.set_facecolor('#efeeea')
plt.bar(range(len(goal_reached)), goal_reached, color="#0A047A", width=1.0)
plt.show()
```

Agent has taken more time to correctly reached to the goal. But, the q table has more non-zero values that the previous. that means, our agent has tried new state, action pairs (explorartion)

```python
##### testing

test_episodes = 100
num_success = 0

for _ in range(test_episodes):
    state = env.reset()
    done = False
    
    while not done:
        action = np.argmax(qtable[state]) 
        
        new_state, reward, done, info = env.step(action)
        state = new_state
        
        num_success += reward

print("Success rate = {} %".format(num_success/test_episodes*100) )
```

```python
# even though it's 100% in this example, more exploration can degrade the performance. But it's neccessary to train 
#the agent that can adapt to new environment.
```

Upto now, we used 'is_slippery=False'. In the slippery environment, the action taken by the agent has only 33% of possibility of succeeding. If a failure, one of the other 3 actions are taken. this adds a more randomness to the training and more difficult to the agent.

```python
######## Training
env = gym.make("FrozenLake-v1", is_slippery=True)
env.reset()

qtable = np.zeros((env.observation_space.n, env.action_space.n))

episodes = 1000
alpha = 0.5       
gamma = 0.9  
epsilon = 1               # amount of randomness
epsilon_decay = 0.001     # linear decay - fixed amount is decreased

goal_reached = []

print("Q table before training :", qtable)

for _ in range(episodes):
    state = env.reset()
    done = False
    goal_reached.append("failure")   
    
    #until the agent reaches to the goal, or gets stuck in a hole
    while not done:
        random_num = np.random.random()
        
        if random_num > epsilon:
            action = np.argmax(qtable[state])    # explotation
        else:
            action = env.action_space.sample()   # exploration
        
        # action
        new_state, reward, done, info = env.step(action)
        # update q
        qtable[state, action] = qtable[state, action] + alpha*(reward + gamma*np.max(qtable[new_state]) - qtable[state, action])
        # update the current state
        state = new_state
        
        if reward:
            goal_reached[-1] = "success"
    
    # linear decay - reduce the epsilon
    epsilon = max(epsilon - epsilon_decay, 0)
    
            
        
print("\nQ table after training :", qtable)
    
# plot if the agent reached the goal in each episode
plt.figure(figsize=(12, 5))
plt.xlabel("episodes")
plt.ylabel("If the agent reached the goal")
ax = plt.gca()
ax.set_facecolor('#efeeea')
plt.bar(range(len(goal_reached)), goal_reached, color="#0A047A", width=1.0)
plt.show()



######### Tetsing

test_episodes = 100
num_success = 0

for _ in range(test_episodes):
    state = env.reset()
    done = False
    
    while not done:
        action = np.argmax(qtable[state]) 
        
        new_state, reward, done, info = env.step(action)
        state = new_state
        
        num_success += reward

print("Success rate = {} %".format(num_success/test_episodes*100) )
```
